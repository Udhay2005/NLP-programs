```python
# ============================================================
# PROGRAM 1: CHATBOT - Rule-Based Conversational Bot
# ============================================================

import re
import random
from datetime import datetime

# --- Pattern-Response pairs ---
patterns = [
    (r'hello|hi|hey|good morning|good evening',
     ['Hello! How can I help you today?',
      'Hi there! What can I do for you?',
      'Hey! Nice to meet you. How can I assist?']),

    (r'how are you|how do you do|are you ok',
     ['I am doing great, thank you for asking!',
      'I am fine! Ready to assist you.',
      'All systems running perfectly! How about you?']),

    (r'what is your name|who are you|your name',
     ['I am ChatBot, your virtual assistant!',
      'My name is ChatBot. I am here to help you!']),

    (r'bye|goodbye|see you|take care',
     ['Goodbye! Have a wonderful day!',
      'Bye! Come back anytime.',
      'See you later! Take care!']),

    (r'thank you|thanks|thank u',
     ['You are welcome!',
      'Happy to help!',
      'Anytime! That is what I am here for.']),

    (r'weather|temperature|climate|forecast',
     ['I cannot check live weather, but try weather.com!',
      'For accurate forecast, please visit a weather site.']),

    (r'time|what time|current time',
     [f'Current time is: {datetime.now().strftime("%H:%M:%S")}',
      f'The time right now is {datetime.now().strftime("%I:%M %p")}']),

    (r'date|today|what day',
     [f'Today is {datetime.now().strftime("%A, %B %d, %Y")}']),

    (r'help|what can you do|features|capabilities',
     ['I can: answer greetings, tell time/date, '
      'chat with you, and respond to common questions!']),

    (r'joke|funny|make me laugh',
     ['Why do programmers prefer dark mode? Because light attracts bugs!',
      'Why did the computer go to the doctor? It had a virus!',
      'I told my computer I needed a break. Now it keeps sending me Kit-Kat ads!']),

    (r'age|how old are you|when were you created',
     ['I am ageless! I was created to assist you anytime.']),

    (r'who made you|who created you|developer',
     ['I was created as part of a Natural Language Processing assignment!']),
]

def chatbot_response(user_input):
    user_input = user_input.lower().strip()
    for pattern, responses in patterns:
        if re.search(pattern, user_input):
            return random.choice(responses)
    return ('I am not sure how to respond to that. '
            'Try asking about time, date, jokes, or just say hello!')

# --- Simulate a conversation ---
test_inputs = [
    'Hello there!',
    'How are you?',
    'What is your name?',
    'Tell me a joke',
    'What is the time?',
    'What is today date?',
    'What can you do?',
    'Who created you?',
    'Thank you so much!',
    'What is quantum physics?',
    'Bye!'
]

print('=' * 55)
print('         CHATBOT CONVERSATION LOG')
print('=' * 55)
for user_msg in test_inputs:
    response = chatbot_response(user_msg)
    print(f'User : {user_msg}')
    print(f'Bot  : {response}')
    print('-' * 55)
```

```python
# ============================================================
# PROGRAM 2: QUESTION ANSWERING using HuggingFace Transformers
# ============================================================

!pip install transformers torch -q

from transformers import pipeline

# --- Load QA Pipeline ---
print("Loading Question Answering model...")
qa_pipeline = pipeline(
    "question-answering",
    model="deepset/roberta-base-squad2"
)
print("Model loaded successfully!\n")

# --- Context Paragraphs ---
context1 = """
Artificial Intelligence (AI) is the simulation of human intelligence
processes by computer systems. It was founded as an academic discipline
in 1956 by John McCarthy. AI applications include expert systems,
natural language processing, speech recognition, and machine vision.
The term AI was coined at the Dartmouth Conference in 1956.
"""

context2 = """
The Python programming language was created by Guido van Rossum
and first released in 1991. Python is known for its simplicity
and readability. It supports multiple programming paradigms including
procedural, object-oriented, and functional programming.
Python is widely used in data science, AI, and web development.
"""

context3 = """
The solar system consists of the Sun and eight planets.
The planets are Mercury, Venus, Earth, Mars, Jupiter, Saturn,
Uranus, and Neptune. Earth is the third planet from the Sun
and the only known planet to support life. Jupiter is the
largest planet in the solar system.
"""

# --- Questions ---
qa_pairs = [
    (context1, 'Who founded Artificial Intelligence?'),
    (context1, 'What are AI applications?'),
    (context1, 'When was the term AI coined?'),
    (context2, 'Who created Python?'),
    (context2, 'When was Python first released?'),
    (context2, 'What is Python used for?'),
    (context3, 'How many planets are in the solar system?'),
    (context3, 'Which planet is the largest?'),
    (context3, 'Which planet supports life?'),
]

print('=' * 62)
print('          QUESTION ANSWERING RESULTS')
print('=' * 62)
for i, (ctx, question) in enumerate(qa_pairs, 1):
    result = qa_pipeline(question=question, context=ctx)
    print(f'Q{i}: {question}')
    print(f'    Answer : {result["answer"]}')
    print(f'    Score  : {result["score"]:.4f}')
    print('-' * 62)
```

```python
# ============================================================
# PROGRAM 3: TEXT SUMMARIZATION (No external model needed)
# ============================================================

# No pip install needed — uses only built-in Python!

# --- Extractive Summarizer from scratch ---
import re
from collections import Counter

def extractive_summarize(text, num_sentences=3):
    # Clean and split into sentences
    sentences = re.split(r'(?<=[.!?])\s+', text.strip())
    sentences = [s.strip() for s in sentences if len(s.split()) > 4]

    # Word frequency (ignore stopwords)
    stopwords = {'the','a','an','is','are','was','were','in','on','at',
                 'to','of','and','or','but','it','its','this','that',
                 'with','for','from','by','as','be','been','have','has',
                 'had','will','would','could','should','may','might','do',
                 'does','did','not','no','so','if','than','then','when',
                 'while','after','before','since','about','into','through',
                 'during','including','without','also','their','they','them',
                 'these','those','which','who','whom','what','how','more'}

    words = re.findall(r'\b[a-z]+\b', text.lower())
    word_freq = Counter(w for w in words if w not in stopwords)

    # Score each sentence
    def score_sentence(sent):
        words_in_sent = re.findall(r'\b[a-z]+\b', sent.lower())
        return sum(word_freq.get(w, 0) for w in words_in_sent if w not in stopwords)

    scored = [(score_sentence(s), i, s) for i, s in enumerate(sentences)]
    scored.sort(reverse=True)

    # Pick top N sentences, restore original order
    top = sorted(scored[:num_sentences], key=lambda x: x[1])
    return ' '.join(s for _, _, s in top)

# --- Input Texts ---
texts = [
    {
        'title': 'Artificial Intelligence',
        'text': (
            "Artificial intelligence (AI) is rapidly transforming every industry. "
            "From healthcare to finance, AI systems now assist doctors in diagnosing "
            "diseases, help banks detect fraud, and power recommendation engines "
            "on platforms like Netflix and Amazon. Machine learning, a subset of AI, "
            "allows computers to learn from data without being explicitly programmed. "
            "Deep learning uses neural networks with many layers to recognize patterns "
            "in images, audio, and text with remarkable accuracy. "
            "Despite its promise, AI raises important ethical concerns about privacy, "
            "bias, job displacement, and accountability that researchers and "
            "policymakers must address carefully."
        )
    },
    {
        'title': 'Climate Change',
        'text': (
            "Climate change refers to long-term shifts in global temperatures and "
            "weather patterns. While natural factors have historically influenced "
            "climate, since the 1800s human activities have been the main driver, "
            "primarily due to burning fossil fuels like coal, oil, and gas. "
            "Rising temperatures cause glaciers to melt, sea levels to rise, "
            "and extreme weather events to become more frequent and intense. "
            "Scientists warn that without significant reductions in greenhouse "
            "gas emissions, the consequences for ecosystems, food security, "
            "and human health will be severe and potentially irreversible."
        )
    },
    {
        'title': 'Space Exploration',
        'text': (
            "Space exploration has achieved remarkable milestones since the first "
            "satellite Sputnik was launched in 1957. NASA Apollo missions landed "
            "humans on the Moon in 1969. Today, the International Space Station "
            "orbits Earth continuously, hosting astronauts from many nations. "
            "Private companies like SpaceX and Blue Origin are revolutionizing "
            "access to space with reusable rockets. Mars missions aim to determine "
            "whether the red planet ever harbored life. Future goals include "
            "returning humans to the Moon and sending astronauts to Mars "
            "within the next two decades."
        )
    },
    {
        'title': 'Renewable Energy',
        'text': (
            "Renewable energy sources such as solar, wind, and hydroelectric power "
            "are becoming increasingly important in the global energy landscape. "
            "Solar panels convert sunlight directly into electricity using photovoltaic cells. "
            "Wind turbines harness kinetic energy from moving air to generate power. "
            "Unlike fossil fuels, renewable energy sources produce little to no greenhouse "
            "gas emissions during operation. Many countries are investing heavily in "
            "renewable infrastructure to reduce carbon footprints and combat climate change. "
            "The cost of solar and wind energy has dropped dramatically over the past decade, "
            "making them competitive with traditional energy sources in many regions."
        )
    },
]

print('=' * 68)
print('         EXTRACTIVE TEXT SUMMARIZATION RESULTS')
print('         Method: TF-based Sentence Scoring')
print('=' * 68)

for item in texts:
    summary = extractive_summarize(item['text'], num_sentences=2)
    original_words = len(item['text'].split())
    summary_words  = len(summary.split())
    reduction      = round((1 - summary_words / original_words) * 100, 1)

    print(f"\nTopic     : {item['title']}")
    print(f"Original  : {original_words} words")
    print(f"Summary   : {summary_words} words  ({reduction}% reduction)")
    print(f"Output    : {summary}")
    print('-' * 68)

print("\n✅ Summarization complete!")
```

```python
# ============================================================
# PROGRAM 4: MACHINE TRANSLATION (Fixed for latest Transformers)
# ============================================================

!pip install transformers sentencepiece torch -q

from transformers import MarianMTModel, MarianTokenizer

# --- Load models directly (no pipeline needed) ---
def load_translator(model_name):
    print(f"  Loading {model_name}...")
    tokenizer = MarianTokenizer.from_pretrained(model_name)
    model     = MarianMTModel.from_pretrained(model_name)
    return tokenizer, model

def translate(text, tokenizer, model):
    tokens = tokenizer([text], return_tensors="pt", padding=True)
    output = model.generate(**tokens)
    return tokenizer.decode(output[0], skip_special_tokens=True)

print("Loading translation models...")
tok_fr, mdl_fr = load_translator("Helsinki-NLP/opus-mt-en-fr")
tok_de, mdl_de = load_translator("Helsinki-NLP/opus-mt-en-de")
tok_es, mdl_es = load_translator("Helsinki-NLP/opus-mt-en-es")
print("All models loaded!\n")

# --- Sentences to translate ---
sentences = [
    "Hello, how are you today?",
    "Artificial Intelligence is changing the world.",
    "The weather is beautiful today.",
    "I love learning new programming languages.",
    "Machine learning is a subset of artificial intelligence.",
    "The book on the table is very interesting.",
    "Technology is advancing at a rapid pace.",
]

print('=' * 68)
print('             MACHINE TRANSLATION RESULTS')
print('=' * 68)
for sent in sentences:
    fr = translate(sent, tok_fr, mdl_fr)
    de = translate(sent, tok_de, mdl_de)
    es = translate(sent, tok_es, mdl_es)
    print(f"EN: {sent}")
    print(f"FR: {fr}")
    print(f"DE: {de}")
    print(f"ES: {es}")
    print('-' * 68)

print("\n✅ Translation complete!")
```
